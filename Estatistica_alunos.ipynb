```python
import pandas as pd
from tkinter import Tk
from tkinter.filedialog import askopenfilename
from datetime import datetime, date
```

```python
# usei o arquivo BD_alunos.csv que colocarei no Repository
# o arquivo foi gerado aleatoriamente com o Excel
Tk().withdraw()
arquivo_alunos = askopenfilename()
```

```python
def calcular_idade(data_nascimento):
    today = date.today()
    return today.year - data_nascimento.year - ((today.month, today.day) < (data_nascimento.month, data_nascimento.day))
```

```python
dados_alunos = pd.read_csv(arquivo_alunos, delimiter =';',decimal=',')
dados_alunos.head()
```

```python
# Tratamentos dos dados, de String para Datas
n = len(dados_alunos['Data de Nascimento'])
idades =[]
for i in range(n):
    date_of_birth = datetime.strptime(dados_alunos['Data de Nascimento'].iloc[i], "%d/%m/%Y")
    idades.append(calcular_idade(date_of_birth))
```

```python
# Inserindo a coluna Idades
dados_alunos.insert(loc=5,column='Idade', value=idades)
```

```python
df = dados_alunos.describe()
df = df[1:]
df = round(df[['Idade','Altura','Peso','CRA','Créditos Obtidos','Renda (Salário Mínimo)']],2)
df
```

```python
df_eng = dados_alunos.loc[dados_alunos['Curso']=='Engenharia'].describe()
df_eng = df_eng[1:]
df_eng = round(df_eng[['Idade','Altura','Peso','CRA','Créditos Obtidos','Renda (Salário Mínimo)']],2)
df_eng
```

```python
df_eco = dados_alunos.loc[dados_alunos['Curso']=='Economia'].describe()
df_eco = df_eco[1:]
df_eco = round(df_eco[['Idade','Altura','Peso','CRA','Créditos Obtidos','Renda (Salário Mínimo)']],2)
df_eco
```

```python
# Colocarei esse arquivo no Repository

df.to_excel('Output_Alunos.xlsx')
with pd.ExcelWriter('Output_Alunos.xlsx') as writer:  
    df.to_excel(writer, sheet_name='Estatística Geral')
    df_fem.to_excel(writer, sheet_name='Estatística Mulheres')
    df_masc.to_excel(writer, sheet_name='Estatística Homens')
    df_eng.to_excel(writer, sheet_name='Estatística Engenharia')
    df_eco.to_excel(writer, sheet_name='Estatística Economia')
```
